```python
class Online_Store:
    import numpy as np
    def __init__(self):
        self.n = []
        self.e = []
        self.pn = []
        self.p = []
        self.pid = [11259,25489,75498,68254,16982,32895,25436,55169,35489,661529]
        self.pname = ["Refrigerator","LCD","Microwave","Speaker","Phone","Laptop","PC","Handset","Charger","Washing Machine"]
        self.price= [25000,18000,8900,4200,58300,94000,149900,1000,500,15000]
        self.stock = [80,50,49,68,23,57,46,150,60,30]
        self.cart= []
        self.order = []
        self.total = []
    def signup(self):
        email=input("\nEnter Your Email ID : ")
        if email not in self.e:
            name=input("Enter Your Good Name : ")
            phone_number=int(input("Enter Your Phone Number : "))
            password=input("Enter Your Password")
            self.n.append(name)
            self.e.append(email)
            self.pn.append(phone_number)
            self.p.append(password)
            print("\tAccount Created Successfully!")
        else:
            print("Email ID Already Exists\n")
#################################################################################################################################################################################################################################
    def login(self):
        def add_products():
            product_name=input("\nEnter Product Name")
            if product_name in self.pname:
                quantity= int(input("Number of product you want"))
                ind=self.pname.index(product_name)
                if self.stock[ind] >= quantity:
                    Total = self.price[ind] * quantity
                    self.cart.append((product_name,quantity , Total ))
                    self.stock[ind] -= quantity
                    print(f"{quantity} {product_name} added to your cart.")
                    order_id=np.random.randint(100000,999999)
                    self.order.append(order_id)
                    print()
                else:
                    print("Not Enough Stock!\n")
            else:
                print("Product Not Availaiblle in Store!. Sorry!!\n")
        def cart():
            try:
                print("\nYOUR ITEMS IN CART : ")
                for i in self.cart:
                    print(f"{i[1]} X {i[0]} at {i[2]}")
            except:
                print("No Items in Cart")
            print()
        def orders():
            print("\nOrder ID : ", self.order[0])
            total_amount = 0
            for i in self.cart:
                print("\t",f"{i[1]} X {i[0]} at {i[2]}")
                total_amount += i[2]
            print("\nTotal Amount : ",total_amount)
            self.total.append(total_amount)
            print()
            x=input("Do you Wish to Continue Press Yes / No : ")
            if x=="Yes":
                z=int(input("\nPress 1 for Paytm 2 for Card : "))
                if z==1:
                    print("\nYou are Choosing Paytm")
                    upi=input("Enter Valid UPI ID : ")
                    upipin=int(input("Enter UPI PIN : "))
                    if upi=="100200300@paytm" and upipin==0000:
                        del self.cart
                        t=np.random.randint(10000000,99999999)
                        print(f"{self.total[0]} amount deducted through UPI Payment Bank.")
                        del self.total
                        print("Your Transaction has been Successfull")
                        print("Your Transaction ID : ",t)
                        print()
                    else:
                        print("Incorrect UPI ID and UPI PIN")
                        print()
                elif z==2:
                    print("\nYou are Choosing Card Payment")
                    cardno=int(input("Enter Valid card No. : "))
                    cvv=int(input("Enter Card CVV"))
                    if cardno==123123123123 and cvv==420:
                        otp=np.random.randint(1000,9999)
                        print("Your OTP : ",otp)
                        OTP=int(input("Enter OTP"))
                        if OTP==otp:
                            del self.cart
                            t=np.random.randint(10000000,99999999)
                            print(f"{self.total[0]} amount deducted through UPI Payment Bank.")
                            del self.total
                            print("Your Transaction has been Successfull")
                            print("Your Transaction ID : ",t)
                            print()
                        else:
                            print("Transaction Cancel")
                            print()
                    else:
                        print("Card Number Not Found")
                else:
                    print("No Payment Method Found!")
            else:
                print("Your Transaction Cancel Successfully!")
        email=input("\nEnter your Email ID : ")
        password= input("Enter your Password : ")
        if email in self.e and password in self.p:
            ind=self.e.index(email)
            while True:
                print("\nPress 1 for Add Products")
                print("Press 2 for Go to Cart")
                print("Press 3 for Check Order Status")
                z=int(input("Press a key : "))
                if z==1:
                    add_products()
                elif z==2:
                    cart()
                elif z==3:
                    orders()
                else:
                    print("Exit")
                    break
        else:
            print("Credential Not Found! \n")
        
###############################################################################################################################################################################################################################################################
    def products(self):
        def allproducts():
            for i in range(len(self.pid)):
                print(f"Products {i+1}:")
                print("\t Product ID    = ",self.pid[i])
                print("\t Product Name  = ",self.pname[i])
                print("\t Product Price = ",self.price[i])
                print("\t Product Stock = ",self.stock[i])
                print()
        def addproducts():
            name=input("\nEnter Product Name : ")
            if name not in self.pname:
                id=np.random.randint(10000,99999)
                stock=int(input("Enter Quantity of Product : "))
                price= int(input("Enter per Piece Price : "))
                self.pid.append(id)
                self.pname.append(name)
                self.price.append(price)
                self.stock.append(stock)
                print("Product Add Successfully")
            else:
                print("Product already in Store\n")
                
        def fillstock():
            ID=int(input("\nEnter Product ID : "))
            if ID in self.pid:
                ind=self.pid.index(ID)
                x=int(input("\nHow much Quantity you have to add in a store : "))
                if x>=1:
                    self.stock[ind]+=x
                    print("Stock Add Successfully")
                else:
                    print("Please Enter Valid Number!")
            else:
                print("Product ID not Found!\n")

        while True:
            print("\nPress 1 for Check Avilaible Products")
            print("Press 2 for Add Products in Store")
            print("Press 3 for Fill Stock")
            x=int(input("\nPress a key : "))
            if x==1:
                allproducts()
            elif x==2:
                addproducts()
            elif x==3:
                fillstock()
            else:
                print("Exit")
                break

#################################################################################################################################################################################################################################
#################################################################################################################################################################################################################################
import getpass as g
import numpy as np
OS= Online_Store()
print("\t\t\t\t\t\t\tWelcome to the Library Management System\n")
EID=int(input("Please Enter your Employee ID : "))
EP=g.getpass("Please Enter your Employee Password : ")
if EID==123 and EP=="1111":
    while True:
        print("\nPress 1 for Signup")
        print("Press 2 for Login")
        print("Press 3 for Check Products")
        print("Press 4 for Exit! ")
        x=int(input("Press a Key : "))
        if x==1:
            OS.signup()
        elif x==2:
            OS.login()
        elif x==3:
            OS.products()
        elif x==4:
            print("Exit")
            break
        else:
            print("Wrong Input")
else:
    print("Wong ID and Password, Try Again Later! ")
print("Thanks, Visit Again!")
```
